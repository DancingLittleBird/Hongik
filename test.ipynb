```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import xlrd
import tkinter
```

```python
#필요한 대로 바꾸시면 됩니다~
XLSX_FILENAME = r'2020-12-24 ABE 및 OCT4 untreated 비교 -1.xlsx'
sheetIndex = 0
xls = xlrd.open_workbook(XLSX_FILENAME, on_demand = True)
sheetnames = xls.sheet_names()
```

```python
dfs = {}
for sheetname in sheetnames:
    dfs[sheetname] = pd.read_excel(XLSX_FILENAME, header=1)
    df = dfs[sheetname]
```

```python
priorindex = 0
dfParts = []
dfPartsColumn = []

columnWithNaNValue = dfs[sheetnames[sheetIndex]].columns[dfs[sheetnames[sheetIndex]].isnull().all(axis = 0)].tolist()
columnIndexWithNaNValue = dfs[sheetnames[sheetIndex]].columns.get_indexer(columnWithNaNValue)

for count, index in zip(range(len(columnIndexWithNaNValue)), columnIndexWithNaNValue):
    dfPartsColumn.append(dfs[sheetnames[sheetIndex]].columns[priorindex:index])
    priorindex = index + 1

dfPartsColumn.append(dfs[sheetnames[sheetIndex]].columns[priorindex:])

for i in range(len(dfPartsColumn)):
    dfParts.append(dfs[sheetnames[sheetIndex]].loc[:,dfPartsColumn[i]])
```

```python
for dfPart in dfParts[1:]:
    for i in range(len(dfPart.columns)):
        dfPart.rename(columns = {dfPart.iloc[:,i].name : dfPart.iloc[:,i].name[0:dfPart.iloc[:,i].name.find(".")]}, inplace = True)
```

```python
genotypeCriterions = []
for i in GENOTYPES:
    genotypeCriterions.append(dfParts[0]["Genotype"].map(lambda x: x == i))
```

```python
genderCriterions = []
for i in GENDER:
    genderCriterions.append(dfParts[0]["Gender"].map(lambda x: x == i))
```

```python
GENOTYPES = dfParts[0]["Genotype"].values().index.to_list()

GENDER = (0,1)
GENDERSTRING = ("Male", "Female")
```

```python
Averages = []
for dfPart in dfParts:
    for i in range(len(GENOTYPES)):
        Averages.append(dfPart[genotypeCriterions[i]]["Average"].mean())
    plt.figure(figsize=(18,3))
    plt.bar(GENOTYPES, Averages)
    plt.show()
    Averages.clear()
```

```python
Averages.clear()
for dfPart in dfParts:
    for i in range(len(GENDER)):
        Averages.append(dfPart[genderCriterions[i]]["Average"].mean())
    plt.figure(figsize=(5,3))
    plt.bar(GENDERSTRING, Averages)
    plt.show()
    Averages.clear()
```
